```python
# Python and Pandas Assessment Notebook

import pandas as pd
import numpy as np
from functools import reduce

# Task 1: Create a DataFrame using Pandas
# Create a DataFrame with the following data:
# Name: Alice, Bob, Charlie, David
# Age: 25, 30, 35, 28
# City: New York, San Francisco, Los Angeles, Chicago

# Your code here:
# Task 1: Create a DataFrame using Pandas

data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Age": [25, 30, 35, 28],
    "City": ["New York", "San Francisco", "Los Angeles", "Chicago"]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)
print("\nData types:")
print(df.dtypes)


# Task 2: Row and Column Manipulation
# Drop the 'City' column from the DataFrame created in Task 1

# Your code here:
# Task 2: Drop the 'City' column

df_no_city = df.drop(columns=["City"])

print("\nDataFrame after dropping City column:")
print(df_no_city)



# Task 3: Handling Null Values
# Create a new DataFrame with some null values and fill them

# Your code here:
# Task 3: Create DataFrame with null values

df_null = pd.DataFrame({
    "Name": ["Alice", None, "Charlie"],
    "Age": [25, None, 35],
    "Score": [90, 85, None]
})

print("\nDataFrame with null values:")
print(df_null)

# Fill null values
df_filled = df_null.fillna({
    "Name": "Unknown",
    "Age": 0,
    "Score": 0
})

print("\nDataFrame after filling null values:")
print(df_filled)


# Task 4: GroupBy and Describe
# Using the following DataFrame, group by 'Category' and describe the 'Value' column

df_group = pd.DataFrame({
    'Category': ['A', 'B', 'A', 'B', 'A', 'C'],
    'Value': [10, 20, 15, 25, 30, 35]
})

# Your code here:
# Task 4: GroupBy and Describe

grouped = df_group.groupby("Category")["Value"].describe()

print("\nGrouped statistics by Category:")
print(grouped)



# Task 5: Concatenation and Merging
# Concatenate two DataFrames vertically and merge them horizontally

df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
df3 = pd.DataFrame({'C': [9, 10], 'D': [11, 12]})

# Your code here:
# Task 5: Concatenate and merge DataFrames

df_concat = pd.concat([df1, df2], ignore_index=True)
print("\nAfter vertical concatenation:")
print(df_concat)

df_merged = pd.concat([df_concat, df3], axis=1)
print("\nAfter horizontal merge:")
print(df_merged)



# Task 6: Tuples and Sets
# Create a tuple of fruits and a set of numbers.
# Then, try to add an element to each and observe the difference.

# Your code here:
# Task 6: Tuples and Sets

fruits = ("apple", "banana", "cherry")
numbers = {1, 2, 3, 4, 5}

print("\nTuple:", fruits)
print("Set:", numbers)

# Attempt to modify both
try:
    fruits.append("orange")
except AttributeError as e:
    print("Tuples are immutable:", e)

numbers.add(6)
print("Modified set:", numbers)



# Task 7: Dictionaries
# Create a dictionary of student names and their scores.
# Then, update a student's score and add a new student.

# Your code here:
# Task 7: Dictionaries

students = {
    "Alice": 85,
    "Bob": 90,
    "Charlie": 78
}

# Update existing student
students["Bob"] = 95

# Add new student
students["Diana"] = 88

print("\nUpdated student dictionary:")
print(students)


# Task 8: Functions and Lambda
# Create a function to calculate the square of a number.
# Then, use a lambda function to do the same.

# Your code here:
# Task 8: Functions and Lambda

def square(x):
    return x * x

square_lambda = lambda x: x * x

print("\nFunction square(4):", square(4))
print("Lambda square(7):", square_lambda(7))



# Task 9: Iterators and Generators
# Create an iterator for the first 5 even numbers.
# Then, create a generator for the same.

# Your code here:
# Task 9: Iterator

class EvenIterator:
    def __init__(self):
        self.current = 0
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < 5:
            value = self.current
            self.current += 2
            self.count += 1
            return value
        else:
            raise StopIteration

print("\nIterator output:")
for num in EvenIterator():
    print(num, end=" ")

# Generator version

def even_generator():
    for i in range(0, 10, 2):
        yield i

print("\n\nGenerator output:")
for num in even_generator():
    print(num, end=" ")



# Task 10: Map, Reduce, and Filter
# Use map to square all numbers in a list.
# Use reduce to find the product of all numbers in a list.
# Use filter to get only even numbers from a list.

numbers = [1, 2, 3, 4, 5]

# Your code here:
# Task 10: Map, Reduce, Filter

squared = list(map(lambda x: x**2, numbers))
product = reduce(lambda x, y: x * y, numbers)
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))

print("\n\nSquared numbers:", squared)
print("Product of numbers:", product)
print("Even numbers:", even_numbers)



# Task 11: Object-Oriented Programming - Creating a Class
# Create a class called 'Rectangle' with attributes 'length' and 'width'.
# Include methods to calculate area and perimeter.

# Your code here:
# Task 11: Rectangle Class

class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

rect1 = Rectangle(5, 3)
rect2 = Rectangle(10, 4)

print("\nRectangle 1 area:", rect1.area())
print("Rectangle 1 perimeter:", rect1.perimeter())
print("Rectangle 2 area:", rect2.area())
print("Rectangle 2 perimeter:", rect2.perimeter())


# Task 12: Pandas Data Analysis
# Using the following DataFrame, perform these tasks:
# 1. Find the average salary by department
# 2. Get the names of employees with salary > 60000
# 3. Add a new column 'Bonus' which is 10% of salary

df_employees = pd.DataFrame({
    'Name': ['John', 'Jane', 'Bob', 'Alice', 'Charlie'],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR'],
    'Salary': [55000, 65000, 70000, 60000, 58000]
})

# Your code here:
# Task 12: Pandas Data Analysis

# Average salary by department
avg_salary = df_employees.groupby("Department")["Salary"].mean()
print("\nAverage salary by department:")
print(avg_salary)

# Employees earning more than 60000
high_earners = df_employees[df_employees["Salary"] > 60000]["Name"]
print("\nEmployees with salary > 60000:")
print(high_earners)

# Add Bonus column
df_employees["Bonus"] = df_employees["Salary"] * 0.10

print("\nEmployee DataFrame with Bonus:")
print(df_employees)
```
